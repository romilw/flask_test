# Machine Learning Model

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/ingledarshan/upGrad_Darshan/main/hiring.csv")
df
```

```python
# I am creating a dummy project project so that we can deploy this
# Salary prediction
# Linear Regression
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.experience.fillna(0,inplace=True)
df.test_score.fillna(df.test_score.mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
def conv_dict(x):
    dict={
        "five":5,
        "two":2,
        "seven":7,
        "three":3,
        "ten":10,
        "eleven":11,
        0:0
    }
    return dict[x]
```

```python
df.experience = df.experience.apply(conv_dict)
```

```python
df
```

# ML Model

```python
X = df.iloc[:,:-1]
X
```

```python
y = df.iloc[:,-1]
y
```

# I am not doing train and test split here, but in real ML model make sure you do it 

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr
```

```python
lr.fit(X,y)
```

```python
lr.predict(X)
```

```python
lr.predict([[2,9,6]])
```
